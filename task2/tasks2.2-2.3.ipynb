```python
import pandas as pd
```

```python
users = pd.DataFrame({
    'id': [1, 2, 3, 4, 5, 6]
})

links = pd.DataFrame({
    'id1': [1, 2, 4],
    'id2': [2, 3, 5]
})

mapping = dict(zip(links['id2'], links['id1']))

def get_root(user_id):
    while user_id in mapping:
        user_id = mapping[user_id]
    return user_id

users['new_id'] = users['id'].apply(get_root)

print('users:')
print(users)

print('\nlinks:')
print(links)
```

```python
users = pd.DataFrame({
    'id': [1, 2, 3, 4, 5, 6]
})

links = pd.DataFrame({
    'id1': [1, 2, 4],
    'id2': [2, 3, 5]
})

mapping = dict(zip(links['id2'], links['id1']))

def get_root(user_id):
    if user_id not in mapping:
        return user_id
    return get_root(mapping[user_id])

users['new_id'] = users['id'].apply(get_root)

users['new_id'] = users['id'].apply(get_root)

print('users:')
print(users)

print('\nlinks:')
print(links)
```

```python
df = pd.DataFrame({
    'id': [1, 2, 3, 4, 5, 6],
    'name': [
        'POMA',
        'РОМА',
        'Рома',
        'Анна',
        'Aнна',
        'АННА'
    ]
})

print(df)

trans = str.maketrans({
    'A': 'А',
    'a': 'а',
    'B': 'В',
    'C': 'С',
    'c': 'с',
    'E': 'Е',
    'e': 'е',
    'H': 'Н',
    'K': 'К',
    'M': 'М',
    'O': 'О',
    'o': 'о',
    'P': 'Р',
    'p': 'р',
    'T': 'Т',
    'X': 'Х',
    'x': 'х'
})

df['clean_name'] = df['name'].str.translate(trans).str.lower()

df['new_id'] = df.groupby('clean_name')['id'].transform('min')

print(df)
```
